```python
import pandas as pd
import json

import mapboxgl
import os

import plotly.express as px
import plotly.io as pio

import plotly.offline as pyo
#plotly offline 모드 변경
pyo.init_notebook_mode()
```

```python
workdir = 'working directory'
filepath1 = '{}/HPC_CNT.dta'.format(workdir)

df = pd.read_stata(filepath1)

df = df.rename(columns={'PROVIENCE_cd':'CTPRVN_CD'})
```

```python
filepath3 = '{}/rok0_008.zip.geojson'.format(workdir)
state_geo2 = json.load(open(filepath3,encoding='utf-8'))

# state_geo2
```

```python
# carto-positron에 mapping해 보았으나 색이 칠해지지 않는 지역이 있음. finetunnig이 필요한듯....
fig = px.choropleth_mapbox( df
                           ,geojson=state_geo2
                           ,locations='CTPRVN_CD'
                           ,color='CNT'
                           ,color_continuous_scale='bluyl'
#                            ,range_color=(0,1)
                           ,mapbox_style="carto-positron"#"open-street-map"#
                           ,featureidkey='properties.CTPRVN_CD'
                           ,zoom=5.7
                           ,center={"lat":36.4, "lon":127.8}
                           ,opacity=1
                           ,labels={'CNT':'수진인원'}
                          )
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(title=dict(text="<b>2020년 지역별 수진현황</b>"
                            ,font_size=40
                            ,x=0.5
                            ,y=0.9
                            )
                 )#, **layout_setting)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
# fig.write_image( '{}/지역별수진현황_sample_mint.png'.format(workdir)
#                 , width=900
#                 , height=600
#                )
# pio.write_image( fig
#                 ,"{}/지역별수진현황_01전국_sample.png".format(workdir)
#                 ,width=1000, height=600
#                 ,scale=1.5
#                )
# fig.show()
```

```python
fig = px.choropleth( df
                    ,geojson=state_geo2
                    ,locations='PROVIENCE'
                    ,color='CNT'
                    ,color_continuous_scale='bluyl'
                    ,featureidkey='properties.CTP_KOR_NM'
                    ,labels={'CNT':'수진인원'}
                    )
# 이거 설정해 주지 않으면 세계 지도 나와버림.
fig.update_geos( fitbounds="geojson"
                ,visible=False
                ,projection=dict(type="mercator" # 지도를 보여주는 방식. mercator, miller, orthographic type recommend
                                ,scale=1.5
                                )
#                 ,center=dict(lat=37, lon=126.986)
               )

fig.update_layout(margin={"r":30,"t":30,"l":30,"b":30})
# fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(title=dict(text="<b>2020년 지역별 수진현황</b>"
                            ,font_size=40
                            ,x=0.5
                            ,y=0.95
                            )
                 )#, **layout_setting)
# fig.write_image는 scale 조정이 안 됨.

# pio.write_image( fig
#                 ,"{}/지역별수진현황_01전국_sample2.png".format(workdir)
#                 ,width=1000, height=600
#                 ,scale=1.5
#                )
fig.show()
```
